```python
import pandas as pd
import numpy as np
import glob

SOURCE = r'../source/'
```

```python
files = glob.glob(pathname= SOURCE + '*.csv')
dfs = {f.split('_')[-1].replace('.csv', ''): pd.read_csv(f) for f in files}
```

```python
dfs.keys()
```

```python
tabelaVeiculo = dfs['tabelaVeiculo']
pois = dfs['pois']
agency = dfs['agency']
tabelaLinha = dfs['tabelaLinha']
pontosLinha = dfs['pontosLinha']
trechosItinerarios = dfs['trechosItinerarios']
linhas = dfs['linhas']
shapeLinha = dfs['shapeLinha']
```

```python
urbs = agency[agency['agency_name']=='Urbanização de Curitiba S/A']
urbs_id = urbs['agency_id'][0].astype('int16')
```

## routes.txt (linhas)

```python
linhas['agency_id']= urbs_id
linhas['route_type'] = np.int16(3)

linhas = linhas.rename(columns={
   'COD':'route_id',
   'NOME':'route_long_name', 
   'NOME_COR':'route_color_name',
   'SOMENTE_CARTAO':'only_card',
   'CATEGORIA_SERVICO':'service_category',
})

linhas = linhas[['route_id', 'agency_id',
                 'route_long_name', 'route_type', 
                 'only_card', 'service_category',
                 'route_color_name']]

linhas.head(5)
```

---


## pois.txt (extra file besides the required ones)

```python
if 'POI_DESC' in pois.columns.values:
    pois = pois.drop(['POI_DESC'], axis=1)

pois = pois.rename(columns={
    'POI_NAME': 'poi_name',
    'POI_CATEGORY_NAME': 'poi_category',
    'POI_LAT': 'poi_lat',
    'POI_LON': 'poi_lon'
})

pois.head(5)
```

---

## stops.txt (pontosLinha)

```python
pontosLinha = pontosLinha.rename(columns={
    'NOME':'stop_name',
    'LAT':'stop_lat',
    'LON':'stop_lon',
    'NUM':'stop_id',
    'TIPO':'stop_desc',
    'SEQ':'seq',
    'GRUPO':'grupo',
    'SENTIDO':'sentido',
    'ITINERARY_ID': 'itinerary_id',
    'COD':'route_id'
})

pontosLinha.head(5)
```
